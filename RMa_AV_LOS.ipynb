```
import numpy as np
import math
import matplotlib.pyplot as plt
```

```
# 調べる距離[m]
d = np.linspace(100,10000,100)
```

```
#ドローンの高さ[m]
hut=150
```

```
#無線周波数[GHz]
fc=2.400
```

```
# ドローンの設定
# MS Tx[dBm]
MsTx = 23  #送信電力
# MS ANT[dBi]
MsANT = 0  #送信アンテナゲイン
```

```
# 基地局
BsANT = 16  #受信アンテナゲイン[dBi]
BsFLos = 3  #フィーダーロス[dB]
```

RMa-AV

A*LOG10(d) + B

```
temp=23.9-1.8*math.log10(hut)
print(temp)
```

```
A = max([temp,20])
print(A)
```

```
B = 20*math.log10((40*math.pi*fc)/3)
```

```
PL = A * np.log10(d) + B
```

```
# 伝搬ロス
plt.plot(d,PL)
```

```
# 受信電力
POW = MsTx + MsANT - PL + BsANT - BsFLos
```

```
# 受信電力のグラフ
plt.plot(d,POW)

plt.xscale("log")
plt.xlabel("d[m]")

plt.grid(which='major',color='black',linestyle='-')
plt.grid(which='minor',color='black',linestyle='-')
plt.show()
```
